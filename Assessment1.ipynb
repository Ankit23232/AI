Task-3 - Outlier detection


```python
#Identify outlier countries in South Asia based on their Score and GDP per Capita.
score = south_asian_dataframe['score']
gdp = south_asian_dataframe['Log GDP per capita']
```

```python
#Define outliers using the 1.5 Ã— IQR rule.
Q1_score = score.quantile(0.25)
Q3_score = score.quantile(0.75)
IQR_score = Q3_score - Q1_score
print('Q1 = ', Q1_score)
print('Q3 = ', Q3_score)
print('IQR = ', IQR_score)
```

```python
low_score = Q1_score -1.5*IQR_score
upper_score = Q3_score+1.5*IQR_score

outlier_score = south_asian_dataframe[(south_asian_dataframe['score']<low_score) | (south_asian_dataframe['score']>upper_score)][['Country name']]
outlier_score

Q1_gdp = gdp.quantile(0.25)
Q3_gdp = gdp.quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

lower_limit_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_limit_gdp = Q3_gdp + 1.5 * IQR_gdp

outlier_gdp = south_asian_dataframe[(south_asian_dataframe['Log GDP per capita']<lower_limit_gdp) | (south_asian_dataframe['Log GDP per capita']>upper_limit_gdp)][['Country name']]

print('low score = ', low_score)
print('Upper score = ', upper_score)
print('Lower limit gdp = ', lower_limit_gdp)
print('Upper limit gdp = ', upper_limit_gdp)
print('Outlier based in gdp = ', outlier_gdp)
print('Outlier based in score= ',outlier_score)
```

```python
#Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
color = ['red' if x in outlier_score['Country name'].values else 'blue' for x in south_asian_dataframe['Country name']]
sns.scatterplot(data = south_asian_dataframe,x='Log GDP per capita',y='score',marker='o',color=color)
plt.grid(True)
plt.title('Score and Log GDP per capita')
plt.xlabel('Log GDP per capita')
plt.ylabel('Score')
```

6. Outlier Detection:

```python
#Identify outlier countries in South Asia based on Score and GDP per Capita.
#Plot these outliers and discuss their implications.

Q1_score_sa = south_asian_dataframe['score'].quantile(0.25)
Q3_score_sa = south_asian_dataframe['score'].quantile(0.75)
IQR_score_sa = Q3_score_sa - Q1_score_sa
lower_bound_score_sa = Q1_score_sa - 1.5 * IQR_score_sa
upper_bound_score_sa = Q3_score_sa + 1.5 * IQR_score_sa

sa_outliers_score = south_asian_dataframe[(south_asian_dataframe['score'] < lower_bound_score_sa) |
                                           (south_asian_dataframe['score'] > upper_bound_score_sa)]

Q1_gdp_sa = south_asian_dataframe['Log GDP per capita'].quantile(0.25)
Q3_gdp_sa = south_asian_dataframe['Log GDP per capita'].quantile(0.75)
IQR_gdp_sa = Q3_gdp_sa - Q1_gdp_sa
lower_bound_gdp_sa = Q1_gdp_sa - 1.5 * IQR_gdp_sa
upper_bound_gdp_sa = Q3_gdp_sa + 1.5 * IQR_gdp_sa

sa_outliers_gdp = south_asian_dataframe[(south_asian_dataframe['Log GDP per capita'] < lower_bound_gdp_sa) |
                                         (south_asian_dataframe['Log GDP per capita'] > upper_bound_gdp_sa)]

sa_gdp_per_capita = pd.concat([sa_outliers_score, sa_outliers_gdp]).drop_duplicates()


plt.figure(figsize=(10, 6))
sns.scatterplot(data=south_asian_dataframe, x='Log GDP per capita', y='score', color='blue', label='South Asia Data')
sns.scatterplot(data=sa_outliers_score, x='Log GDP per capita', y='score', color='red', label='Outliers (Score)')
sns.scatterplot(data=sa_gdp_per_capita, x='Log GDP per capita', y='score', color='orange', label='Outliers (GDP per Capita)')
plt.title('Outliers in South Asia: Score vs GDP per Capita')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```
