```python
import pandas as pd
import numpy as np

# Load data
data = pd.read_csv('/content/Titanic-Dataset.csv')

# Assuming the features (X) and target (Y) columns are as follows:
# Replace 'feature1', 'feature2', ..., 'target' with the actual column names
X = data[['Survived', 'Age']].values  # Replace with actual feature columns
Y = data['Name'].values  # Replace with the actual target column

# Custom train_test_split function
def train_test_split(X, Y, test_size=0.2, random_state=None):
    """
    Splits the dataset into training and testing sets.

    Parameters:
        X: Feature Matrix (numpy array or similar)
        Y: Target Vector (numpy array or similar)
        test_size: Proportion of the dataset to include in the test split (default: 0.2)
        random_state: Seed for reproducibility (default: None)
    Returns:
        X_train, X_test, Y_train, Y_test: Split datasets
    """
    if random_state is not None:
        np.random.seed(random_state)

    # Shuffle indices
    indices = np.arange(len(X))
    np.random.shuffle(indices)

    # Shuffle X and Y
    X_shuffled, Y_shuffled = X[indices], Y[indices]

    # Calculate split index
    split_index = int(len(X) * (1 - test_size))

    # Split data
    X_train, X_test = X_shuffled[:split_index], X_shuffled[split_index:]
    Y_train, Y_test = Y_shuffled[:split_index], Y_shuffled[split_index:]

    return X_train, X_test, Y_train, Y_test

# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Output Results
print("Training Data Shape:", X_train.shape, Y_train.shape)
print("Testing Data Shape:", X_test.shape, Y_test.shape)

print("\nX_train:\n", X_train)
print("\nX_test:\n", X_test)
print("\nY_train:\n", Y_train)
print("\nY_test:\n", Y_test)
```

```python
def cost_function(X, Y, W):
    """
    Calculates the Mean Squared Error (MSE)
    Parameters:
        X: Feature Matrix
        Y: Target Vector
        W: Weight Vector
    Returns:
        cost: Mean Squared Error
    """
    m = len(Y)
    predictions = np.dot(X, W)
    cost = (1 / (2 * m)) * np.sum((predictions - Y) ** 2)
    return cost
```

```python
# Test case
X_test_case = np.array([[1, 2], [3, 4], [5, 6]])
Y_test_case = np.array([3, 7, 11])
W_test_case = np.array([1, 1])

# Calculate cost
cost = cost_function(X_test_case, Y_test_case, W_test_case)
print("Test Case Cost:", cost)
```

```python
def gradient_descent(X, Y, W, learning_rate, iterations):
    """
    Performs Gradient Descent to optimize weights.
    Parameters:
        X: Feature Matrix
        Y: Target Vector
        W: Initial Weights
        learning_rate: Learning Rate
        iterations: Number of iterations
    Returns:
        W: Optimized Weights
        cost_history: List of cost function values
    """
    m = len(Y)
    cost_history = []

    for i in range(iterations):
        predictions = np.dot(X, W)
        gradients = (1 / m) * np.dot(X.T, (predictions - Y))
        W = W - learning_rate * gradients

        cost = cost_function(X, Y, W)
        cost_history.append(cost)

        if i % 100 == 0:
            print(f"Iteration {i}, Cost: {cost}")

    return W, cost_history
```

```python
# Add a bias column (all 1s) to X_train
X_train_bias = np.hstack([np.ones((X_train.shape[0], 1)), X_train])
X_test_bias = np.hstack([np.ones((X_test.shape[0], 1)), X_test])

# Initialize weights (W) as zeros
W_initial = np.zeros(X_train_bias.shape[1])

# Set learning rate and iterations
survived_rate = 0.01
iterations = 1000
```

```python
# Train the model
if isinstance(Y_train, (list, np.ndarray)) and isinstance(Y_train[0], str):
    Y_train = np.array([1 if y == "yes" else 0 for y in Y_train])

# Verify that Y_train is now numerical
Y_train = np.array(Y_train, dtype=float)

# Example gradient descent function
def gradient_descent(X, Y, W, learning_rate, iterations):
    m = len(Y)  # Number of training examples
    cost_history = []

    for i in range(iterations):
        predictions = np.dot(X, W)
        gradients = (1 / m) * np.dot(X.T, (predictions - Y))
        W = W - learning_rate * gradients

        # Compute cost for monitoring (optional)
        cost = (1 / (2 * m)) * np.sum((predictions - Y) ** 2)
        cost_history.append(cost)

    return W, cost_history

# Now call the function
W_optimal, cost_history = gradient_descent(X_train_bias, Y_train, W_initial, survived_rate, iterations)

print("Optimized Weights:", W_optimal)
```

```python
# Predictions
Y_pred_train = np.dot(X_train_bias, W_optimal)
Y_pred_test = np.dot(X_test_bias, W_optimal)

# Plot cost history
plt.plot(cost_history)
plt.title("Cost Function History")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Remove NaN rows from training data
valid_train = ~np.isnan(Y_train) & ~np.isnan(Y_pred_train)
Y_train = Y_train[valid_train]
Y_pred_train = Y_pred_train[valid_train]

# Remove NaN rows from testing data
valid_test = ~np.isnan(Y_test) & ~np.isnan(Y_pred_test)
Y_test = Y_test[valid_test]
Y_pred_test = Y_pred_test[valid_test]
# Evaluate on training and testing sets
mse_train = mean_squared_error(Y_train, Y_pred_train)
mse_test = mean_squared_error(Y_test, Y_pred_test)

r2_train = r2_score(Y_train, Y_pred_train)
r2_test = r2_score(Y_test, Y_pred_test)

print("Training MSE:", mse_train)
print("Testing MSE:", mse_test)
print("Training R2 Score:", r2_train)
print("Testing R2 Score:", r2_test)
```
