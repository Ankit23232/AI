```python
def catAge(age):
    if age <18:
      return 'child'
    elif age >=18 and age <65:
      return 'adult'
    else:
      return 'senior'
dataframe['age_cat'] = dataframe['Age'].apply(catAge)
print(dataframe.head())
```

```python
totalPassenger = len(dataframe)
print(totalPassenger)
```

```python
Child = dataframe[dataframe['age_cat'] == 'child']
child_count = len(Child)

Adult = dataframe[dataframe['age_cat'] == 'adult']
adult_count = len(Adult)

senior = dataframe[dataframe['age_cat'] == 'senior']
senior_count = len(senior)

print(child_count)
print(adult_count)
print(senior_count)
```

```python
def categorize_age(age):
    if age < 18:
        return 'child'
    elif 18 <= age < 65:
        return 'adult'
    elif age >= 65:
        return 'senior'

dataframe ['age_group'] = dataframe['Age'].apply(categorize_age)
age_group_count = dataframe['age_group'].value_counts()
total_passenger = len(dataframe)
proportion = (age_group_count / total_passenger) * 100

print("In percentage: ", proportion)
```
