```python
x = dataframe.drop(columns = ['Outcome'])
y = dataframe['Outcome']
```

```python
#Perform train-test split (70%-30%)
np.random.seed(5)
indices = np.arange(x.shape[0])
np.random.shuffle(indices)
test_split_size = int(len(x) * 0.3)
test_indices = indices[:test_split_size]
train_indices = indices[test_split_size:]

x_train = x.iloc[train_indices]
X_test = x.iloc[test_indices]

y_train = dataframe['Outcome'].iloc[train_indices]
y_test = dataframe['Outcome'].iloc
[test_indices]


print("Training set size:", x_train.shape, y_train.shape)
print("Testing set size:", X_test.shape, y_test.shape)
```

```python
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))
```

```python
def predict_single(X_train, y_train, query, k=3):
    distances = []

    for i, train_point in enumerate(X_train):
        dist = euclidean_distance(query, train_point)
        distances.append((dist, y_train[i]))

    distances.sort(key=lambda x: x[0])
    k_nearest_neighbors = distances[:k]

    k_labels = [label for _, label in k_nearest_neighbors]

    return max(set(k_labels), key=k_labels.count)
```

```python
def predict_all(X_train, y_train, X_test, k=3):
    predictions = []
    for query in X_test:
        pred = predict_single(X_train, y_train, query, k)
        predictions.append(pred)
    return np.array(predictions)
```

```python
def evaluate_accuracy(y_true, y_pred):
    correct = np.sum(y_true == y_pred)
    total = len(y_true)
    return correct / total
```

```python
k = 3
y_pred = predict_all(x_train, y_train, X_test, k)

accuracy = evaluate_accuracy(y_test, y_pred)

print("Predictions:", y_pred)
print("Accuracy:", accuracy)
```

```python
x_train_np = x_train.to_numpy()
y_train_np = y_train.to_numpy()
X_test_np = X_test.to_numpy()
y_test_np = y_test.to_numpy()

y_pred_custom = predict_all(x_train_np, y_train_np, X_test_np, k=3)
accuracy_custom = evaluate_accuracy(y_test_np, y_pred_custom)

print("Predictions (custom dataset):", y_pred_custom)
print("Accuracy (custom dataset):", accuracy_custom)
```

```python
def scale_features(X):
    mean = x.mean(axis=0)
    stds = X.std(axis=0)

    X_scaled = (X - mean) / stds
    return X_scaled

X_train_scaled = scale_features(x_train)
X_test_scaled = scale_features(X_test)
```

```python
# Use the KNN function to predict on the scaled data
y_pred_scaled = predict_all(X_train_scaled, y_train, X_test_scaled, k=3)

# Evaluate the accuracy on the scaled data
accuracy_scaled = evaluate_accuracy(y_test, y_pred_scaled)

print("Accuracy on scaled data:", accuracy_scaled)
```

```python
y_pred_original = predict_all(x_train, y_train, X_test, k=3)

accuracy_original = evaluate_accuracy(y_test, y_pred_original)

print("Accuracy on original data:", accuracy_original)
```

```python
print(f"Accuracy on Original Data: {accuracy_original}")
print(f"Accuracy on Scaled Data: {accuracy_scaled}")
```

```python
def plot_results(results_no_scaling, results_scaled, metric, ylabel):
    """Helper function to plot results."""
    k_values = [result['k'] for result in results_no_scaling]
    metric_no_scaling = [result[metric] for result in results_no_scaling]
    metric_scaled = [result[metric] for result in results_scaled]

    plt.figure(figsize=(10, 6))
    plt.plot(k_values, metric_no_scaling, label='Without Scaling', marker='o')
    plt.plot(k_values, metric_scaled, label='With Scaling', marker='s')
    plt.xlabel('Number of Neighbors (k)')
    plt.ylabel(ylabel)
    plt.title(f'k vs. {ylabel}')
    plt.legend()
    plt.grid()
    plt.show()

# Plot accuracy
plot_results(accuracy_custom, accuracy_scaled, metric='accuracy', ylabel='Accuracy')

# Plot time taken
plot_results(accuracy_custom, accuracy_scaled, metric='time', ylabel='Time Taken (seconds)')

# Analyze and Discuss
print("--- Analysis and Discussion ---")
print("1. Impact of k on Accuracy:")
print("   - Smaller k values are more sensitive to noise and may lead to overfitting.")
print("   - Larger k values provide smoother decision boundaries but may underfit the data.")

print("2. Impact of k on Computational Cost:")
print("   - As k increases, the computational cost slightly increases due to the sorting and selection of more neighbors.")

optimal_k = max(results_scaled, key=lambda x: x['accuracy'])['k']
print(f"3. Optimal k: Based on the analysis, the optimal k is {optimal_k}, as it provides the highest accuracy.")
```
